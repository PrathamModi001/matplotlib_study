```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
x = np.arange(0,10)
y = 2*x
```

# Functional Method

```python
plt.plot(x,y);
plt.title("A Linear Equation");
plt.xlabel("X Axis");
plt.ylabel("Y Axis");

# Provide Limits: 
plt.xlim(0,)
plt.ylim(0,20)

# Save the plot:
plt.savefig("firstplot.png");
```

# OOP Method: 

```python
# Adds a blank canvas, we'll call this figure
fig = plt.figure(figsize=(5,5))
```

```python
fig = plt.figure(figsize=(5,5))
axes = fig.add_axes([0,0,1,1])

# [0,0, , ] 
# 0,0 will tell us where the origin will be present
# [ , ,1,1]
# 1,1 is saying take 100% width 100% height of the blank canvas(figure)
```

```python
a = np.linspace(0,10,11)
b = a**4
x = np.arange(0,10)
y = 2*x
```

```python
fig = plt.figure() # Create a figure (blank canvas)

axes = fig.add_axes([0,0,1,1]) # Add axes to the figure

axes.plot(a,b) # Plot it
```

```python
fig = plt.figure(dpi=200,figsize=(10,3)) # It is a bit blurry, so do dpi=100, 200

# LARGE Figure
axes = fig.add_axes([0,0,1,1]) # Add axes to the figure
# The change in the syntax btw functional and oop method


axes.set_title("BIG FIGURE")
axes.set_xlabel("X Axis")
axes.set_ylabel("Y Axis")

axes.plot(a,b) # Plot it

# SMALLER Figure
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])
axes2.plot(a,b);
axes2.set_title("Zoomed In")
axes2.set_xlim(6,8)
axes2.set_ylim(1500,4500)
axes2.set_xlabel("X Axis")
axes2.set_ylabel("Y Axis")

#kyuki apnna plot thoda ajeeb size ka hai while viewing after saving kuchh gadbad aa sakti isliye do: 
# Bounding Box = Tight
fig.savefig("New_fig.png",bbox_inches='tight')
plt.show()
```

## Subplots

```python

fig,axes = plt.subplots(nrows=2,ncols=1, dpi=200)
axes[0].plot(x,y)
axes[0].set_title("Plot 1")

axes[1].plot(a,b)

fig.suptitle("Figure Level Title")

# Sometimes they'll be very close and overlapping one another so do: 
plt.tight_layout()
# If still you want custom spacing, you can do: 
# fig.subplots_adjust() wspace, hspace adjust karo 

fig.savefig("subplots.png", dpi=200, bbox_inches="tight")
```

# Styling

## Legend

```python
fig = plt.figure(dpi=200)
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, label="X Vs. X");
ax.plot(x,x**2, label="X Vs. X^2");

ax.legend(loc=(1.1,0.5)) #loc=0 for best automatic placements or loc="lower left"
```

## Changing Colors and Styles

```python
fig = plt.figure(dpi=200)
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color="purple") # Can also pass on RGB Hex Code
```
