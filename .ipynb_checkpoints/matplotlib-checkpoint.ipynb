# Matplotlib Workflow: 

```python
#0 Import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

#1 Prepare Data:
x = [1,2,3,4,5,6]
y = [20,30,40,90,100,80]

#2 Setup Plot
fig, ax = plt.subplots(figsize=(10,10)) # figsize specifies width and height

#3 Plot Data 
ax.plot(x,y)

#4 Customize Plot
ax.set(title="Simple plot",
      xlabel="x-axis",
      ylabel="y-axis")

#5 Save and Export
fig.savefig("./sample-plot.png")
```

# Making plots with NumPy Arrays

Main types of plots: 

* Line Plot
* Scatter Plot
* Bar Plot
* Histograms
* Subplots

```python
import numpy as np
```

```python
# Create some data

x = np.linspace(0,10,100)
y = x ** 2
```

### Line Plot

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,y);
```

### Scatter Plot

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(x,np.sin(x));
```

### Bar Plot

```python
nut_butter_prices = {
    "Almond Butter" : 18,
    "Peanut Butter" : 12,
    "Cashew Butter" : 25    
}
fig, ax = plt.subplots(figsize=(5,5))
ax.bar( nut_butter_prices.keys() , nut_butter_prices.values() );

ax.set(
    title="Nut Butter Prices",
    xlabel="Nuts",
    ylabel="Prices"
);
```

```python
# Horizontal Bar Plot

fig,ax = plt.subplots(figsize=(5,5))
ax.barh( list(nut_butter_prices.keys()) , list(nut_butter_prices.values()) );

# only difference while plotting barh on dictionary: Must convert keys and values to LIST.
```

### Histograms

```python
x = np.random.randn(1000)

fig, ax = plt.subplots(figsize=(5,5))
ax.hist(x);
```

# 2 Options for creating subplots

```python
# Option 1: Create multiple subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2, 
                                             figsize=(10, 5))

# Plot data to each axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Option 2:
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Index to plot data
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

# Plotting from Pandas Dataframes

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '', regex=True)
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2023", periods=len(car_sales))
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales["Total"] = car_sales["Price"].cumsum()
car_sales
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(car_sales["Sale Date"] , car_sales["Total"]);

# Since the dates became unreadable, we will rotate them 90 degrees and show only the date and month
# you can search all this up dont remember
import matplotlib.dates as mdates

date_formatter = mdates.DateFormatter('%m-%d');
plt.gca().xaxis.set_major_formatter(date_formatter);
plt.xticks(rotation=90);

ax.set(title="Modi Car Sales",
      xlabel="Date",
      ylabel="Total Sales");
```

```python
# Another way to plot through Dataframes
car_sales.plot.bar(x="Make" , y="Odometer (KM)");
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.hist(heart_disease["age"],bins=30);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
average = np.mean(over_50["chol"])

fig, ax = plt.subplots(figsize=(10,5))
# Plot
# x,y and c, c here is used to specify that this will be used for colors
scatter = ax.scatter(x=over_50["age"],y=over_50["chol"], c=over_50["target"] );
# Customise
ax.set(title="Heart Disease",xlabel="Age", ylabel="Cholestrol");
ax.legend(*scatter.legend_elements(), title="target");

plt.axhline(average, color='r', linestyle='--', label='Average');
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 15))

# Plotting the 2 Subplots
ax0 = ax[0].scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
ax1 = ax[1].scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])

# plt.subplots_adjust(hspace=0.6)
chol_mean = np.mean(over_50["chol"]);
thalach_mean = np.mean(over_50["thalach"])

# Customising
ax[0].set(title="Age vs Cholestrol",xlabel="Age",ylabel="Cholestrol");
ax[1].set(title="Age vs Max Heart Rate",xlabel="Age",ylabel="Max Heart Rate");

# Legend
ax[0].legend(*ax0.legend_elements(), title="target");
ax[1].legend(*ax0.legend_elements(), title="target");

# Mean Line
ax[0].axhline(chol_mean, color='r', linestyle='--', label='Average');
ax[1].axhline(thalach_mean, color='r', linestyle='--', label='Average');

# Adding an entire figure top heading
fig.suptitle("Heart Disease Analysis", fontsize=20,fontweight="bold");

fig.savefig("./heart-disease.png")
```

```python
# cmap Changes the coloring scheme of the graph:

fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"], 
                     cmap="winter")
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")
ax.axhline(y=over_50["chol"].mean(), 
           color='r', 
           linestyle='--', 
           label="Average");
ax.legend(*scatter.legend_elements(), title="Target");
```

### Changing xlim, ylim

```python
## BEFORE XLIM YLIM
## Before the change (we've had color updates)

fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(10, 10))
scatter = ax0.scatter(over_50["age"], 
                      over_50["chol"], 
                      c=over_50["target"],
                      cmap='winter')
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

# Setup a mean line
ax0.axhline(y=over_50["chol"].mean(), 
            color='r', 
            linestyle='--', 
            label="Average");
ax0.legend(*scatter.legend_elements(), title="Target")

# Axis 1, 1 (row 1, column 1)
scatter = ax1.scatter(over_50["age"], 
                      over_50["thalach"], 
                      c=over_50["target"],
                      cmap='winter')
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Setup a mean line
ax1.axhline(y=over_50["thalach"].mean(), 
            color='r', 
            linestyle='--', 
            label="Average");
ax1.legend(*scatter.legend_elements(), title="Target")

# Title the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');


## AFTER XLIM YLIM

## After adding in different x & y limitations

fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(10, 10))
scatter = ax0.scatter(over_50["age"], 
                      over_50["chol"], 
                      c=over_50["target"],
                      cmap='winter')
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

# Set the x axis
ax0.set_xlim([50, 80])

# Setup a mean line
ax0.axhline(y=over_50["chol"].mean(), 
            color='r', 
            linestyle='--', 
            label="Average");
ax0.legend(*scatter.legend_elements(), title="Target")

# Axis 1, 1 (row 1, column 1)
scatter = ax1.scatter(over_50["age"], 
                      over_50["thalach"], 
                      c=over_50["target"],
                      cmap='winter')
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Set the y axis
ax1.set_ylim([60, 200])

# Setup a mean line
ax1.axhline(y=over_50["thalach"].mean(), 
            color='r', 
            linestyle='--', 
            label="Average");
ax1.legend(*scatter.legend_elements(), title="Target")

# Title the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
